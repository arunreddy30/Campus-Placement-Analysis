```python

import matplotlib.pyplot as plt
plt.figure(figsize=(7, 5))
data_excludenull=data[data['salary_as_fresher']!=0]
grouped_data = data_excludenull.groupby('branch')['salary_as_fresher'].agg(['max', 'mean', 'min'])

# Get the branches and corresponding average, highest, and least salaries
branches = grouped_data.index.tolist()
avg_salaries = grouped_data['mean'].tolist()
highest_salaries = grouped_data['max'].tolist()
least_salaries = grouped_data['min'].tolist()
branch_list={}
for i in range(len(branches)):
  branch_list[branches[i]]={
      'average_sal':np.round(avg_salaries[i]),
      'highest_sal':np.round(highest_salaries[i]),
      'least_sal':np.round(least_salaries[i])
  }

# Set the width of the bars
bar_width = 0.2
# Create a bar plot for highest salary
plt.barh([i for i in range(len(branches))], highest_salaries, height=bar_width, label='Highest Salary', color='green')
for i in range(len(branches)):
    plt.text(highest_salaries[i] + 0.4, i, str(highest_salaries[i]), ha="left", va="center")

# Create a bar plot for average salary
plt.barh([i + bar_width for i in range(len(branches))], avg_salaries, height=bar_width, label='Average Salary', color='blue')
for i in range(len(branches)):
    plt.text(avg_salaries[i] + 0.4, i + bar_width, str(round(avg_salaries[i], 2)), ha="left", va="center")

# Create a bar plot for least salary
plt.barh([i + bar_width * 2 for i in range(len(branches))], least_salaries, height=bar_width, label='Least Salary', color='red')
for i in range(len(branches)):
    plt.text(least_salaries[i] + 0.4, i + bar_width * 2, str(least_salaries[i]), ha="left", va="center")

# Set the y-axis labels
plt.yticks([i + bar_width for i in range(len(branches))], branches)

# Add labels and title
plt.ylabel('Branch')
plt.xlabel('Salary')
plt.title('Highest, Average and Least Salary for Each Branch')
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()


```

##Technical Skills influencing No of projects done by the student

```python
data_pro = pd.DataFrame()
data_pro['no_of_projects']=data['no_of_projects']
data_pro['technical_skills'] = data['web_dev'] + data['Machine Learning'] + data['mobile_dev'] + data['cloud']
data_pro['technical_skills'] = data_pro['technical_skills'].astype(int)
data_pro=data_pro[data_pro['technical_skills'] > 0]
# Calculate correlation coefficient
correlation_coefficient = data_pro[data_pro['technical_skills']>0]['technical_skills'].corr(data_pro['no_of_projects'])

# Create a scatter plot
plt.scatter(data_pro[data_pro['technical_skills']>0]['technical_skills'], data_pro['no_of_projects'], c='blue', alpha=0.7, label='Data Points')

# Add labels and title
plt.xlabel('Technical Skills')
plt.ylabel('Number of Projects')
plt.title(f'Correlation between Number of Projects and Technical Skills (Correlation: {correlation_coefficient:.2f})')

# Calculate the best-fit line (correlation line)
slope, intercept = np.polyfit(data_pro['technical_skills'], data_pro['no_of_projects'], 1)
best_fit_line = slope * data_pro['technical_skills'] + intercept

# Plot the best-fit line
plt.plot(data_pro['technical_skills'], best_fit_line, color='red', label='Correlation Line')
plt.xticks(np.unique(list(data_pro['technical_skills'])))
# Add a legend
plt.legend()

# Show the plot
plt.show()
```

##Percentage of Students Placed and Not Placed in each Branch

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data, replace this with your actual data
data_nested_pie = {
    'branch': data.groupby('branch')['is_placed'].mean().index.tolist(),
    'is_placed':data.groupby('branch')['is_placed'].mean().values,  # Percentage of students placed in each branch
}

# Create a DataFrame from the data
data_nspie = pd.DataFrame(data_nested_pie)

# Function to draw the nested pie chart
def draw_nested_pie_chart(df):
    total_students = 100  # Assuming 100% total students
    total_not_placed = 100 - df['is_placed'] * total_students

    plt.figure(figsize=(12, 6))

    for i, row in df.iterrows():
        branch_name = row['branch']
        placed_percentage = row['is_placed'] * total_students
        not_placed_percentage = total_not_placed[i]

        # Outer pie chart - Percentage of students in each branch
        plt.subplot(1, len(df), i + 1)
        outer_labels = [f'{branch_name} - Placed', f'{branch_name} - Not Placed']
        outer_sizes = [placed_percentage, not_placed_percentage]
        plt.pie(outer_sizes, labels=outer_labels, autopct='%1.1f%%', startangle=90, labeldistance=1)

    plt.suptitle('Nested Pie Chart\nPercentage of Students Placed/Not Placed in Each Branch')
    plt.axis('equal')  # Equal aspect ratio ensures that pies are drawn as circles.
    plt.show()

# Call the function to draw the nested pie chart
draw_nested_pie_chart(data_nspie)
```
